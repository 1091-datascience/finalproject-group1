```python
import pandas
import pandas_datareader as pdr
from datetime import datetime, timedelta
import numpy as np

sd = datetime(2009, 1, 1)
ed = datetime(2021, 1, 8)

vfinx = pdr.get_data_yahoo(symbols='VFINX', start=sd, end=ed)["Adj Close"]
vix = pdr.get_data_yahoo(symbols='^vix', start=sd, end=ed)["Adj Close"]
spx = pdr.get_data_yahoo(symbols='^SP500TR', start=sd, end=ed)["Adj Close"]
vfx = pdr.get_data_yahoo(symbols='VFINX', start=sd, end=ed)["Adj Close"]
vbx = pdr.get_data_yahoo(symbols='VBMFX', start=sd, end=ed)["Adj Close"]
usdx= pdr.get_data_yahoo(symbols='DX-Y.NYB', start=sd, end=ed)["Adj Close"]
tnx = pdr.get_data_yahoo(symbols='^TNX', start=sd, end=ed)["Adj Close"]
nasdaq= pdr.get_data_yahoo(symbols='NQ=F', start=sd, end=ed)["Adj Close"]
dji = pdr.get_data_yahoo(symbols='^DJI', start=sd, end=ed)["Adj Close"]
oil = pdr.get_data_yahoo(symbols='CL=F', start=sd, end=ed)["Adj Close"]
apple = pdr.get_data_yahoo(symbols='AAPL', start=sd, end=ed)["Adj Close"]
msft = pdr.get_data_yahoo(symbols='MSFT', start=sd, end=ed)["Adj Close"]
#fb = pdr.get_data_yahoo(symbols='FB', start=sd, end=ed)["Adj Close"]
google = pdr.get_data_yahoo(symbols='GOOGL', start=sd, end=ed)["Adj Close"]
amazon = pdr.get_data_yahoo(symbols='AMZN', start=sd, end=ed)["Adj Close"]
spyVol = pdr.get_data_yahoo(symbols='SPY', start=sd, end=ed)["Volume"]
```

```python
bnp = pandas.concat([vfinx,vix ,vfx ,vbx ,spx,usdx,tnx,nasdaq,dji,oil,apple,msft,google,amazon,spyVol], axis=1) #state_result is the labels of trend 
bnp = bnp.sort_values(by="Date",ascending=True)
bnp.columns=['VFINX','vix','vfx','vbx','spx','usdx','tnx','nasdaq','dji','oil','apple','msft','google','amazon','spyVol']
bnp.tail(80)
```

```python
bnp = bnp.dropna(subset = ['VFINX'])
```

```python
bnp = bnp.reset_index()
```

# 只留禮拜五的數據

```python
friday_list = pandas.DataFrame()
day_count = 0
for i in range(len(bnp)):
    friday_find = False
    day_count += 1
    if bnp['Date'][i].isoweekday() == 5:
        friday_list = friday_list.append(bnp.iloc[i])
        day_count = 0
        friday_find = True
        continue
    if day_count > 4:
        friday_list = friday_list.append(bnp.iloc[i])
        day_count = 0
```

```python
friday_list = friday_list.set_index('Date')
friday_list = friday_list[['VFINX','vix','vfx','vbx','spx','usdx','tnx','nasdaq','dji','oil','apple','msft','google','amazon','spyVol','state_result']]
```

```python
friday_list
```

# 計算周簡式報酬率 (p2 - p1)/ p1

```python
import numpy as np

for i in range(1,len(friday_list)):
    friday_list['state_result'][i] = (friday_list['VFINX'][i] - friday_list['VFINX'][i-1])/friday_list['VFINX'][i-1]
friday_list = friday_list.reset_index()
friday_list = friday_list.drop(index= 0)
friday_list = friday_list.set_index('Date')
friday_list = friday_list.reset_index()
```

```python
rate_return = friday_list['state_result'].rolling(30).std()
rate_return
```

```python
friday_list
```

```python
friday_list['state_result'][0]
```

# 計算漲 跌 持平標籤

```python
for i in range(30, len(rate_return)):
    if friday_list['state_result'][i] > rate_return[i-1]:
        friday_list.loc[i,'state_result'] = 2 #漲
    elif friday_list['state_result'][i] <= rate_return[i-1] and friday_list['state_result'][i] >= rate_return[i-1] * (-1):
        friday_list.loc[i,'state_result'] = 1 #持平
    elif friday_list['state_result'][i] < rate_return[i-1] * (-1):
        friday_list.loc[i,'state_result'] = 0 #跌
```

```python
friday_list = friday_list.set_index('Date')
friday_list
```

```python
bnp.to_csv('Result.csv')
#friday_list.to_csv('friday.csv')
```

```python
import requests
res = requests.get('https://news.ltn.com.tw/list/breakingnews')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text,'lxml')
```

```python
newsary = []
for li in soup.select('ul.list li'):
  #print(li)
  title = li.select_one('.title').text
  dt = li.select_one('.time').text
  link = li.select_one('a').get('href')
  newsary.append({'title':title, 'time': dt, 'link':link})
```

```python
import pandas
newsdf = pandas.DataFrame(newsary)
newsdf.head()
```

```python
import requests
res = requests.get('https://news.ltn.com.tw/ajax/breakingnews/all/4')
jd = res.json()
import pandas
df = pandas.DataFrame(jd['data'].values())
import time
newsary = []
for i in range(2,5):
  res = requests.get('https://news.ltn.com.tw/ajax/breakingnews/all/{}'.format(i))
  time.sleep(2)
  #print(res.json())
  jd = res.json()
  df = pandas.DataFrame(jd['data'].values())
  newsary.append(df)
```

```python
newsdf = pandas.concat(newsary)
```
